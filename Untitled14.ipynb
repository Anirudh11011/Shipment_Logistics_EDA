```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)
sns.set_theme(style="whitegrid")

RANDOM_STATE = 42
```

## 1) LOAD DATA

```python
DATA_PATH = "smart_logistics_dataset.csv"   # <-- change if needed
df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
display(df.head())
```

## 2) BASIC DATA AUDIT (TYPES, NULLS, DUPLICATES)

```python
def audit_df(data: pd.DataFrame) -> pd.DataFrame:
    out = pd.DataFrame({
        "dtype": data.dtypes.astype(str),
        "null_count": data.isna().sum(),
        "null_%": (data.isna().mean() * 100).round(2),
        "n_unique": data.nunique(dropna=True),
    })
    return out.sort_values(["null_%", "n_unique"], ascending=False)

display(audit_df(df))

dup_count = df.duplicated().sum()
print("Duplicate rows:", dup_count)
```

## 3) STANDARDIZE TYPES + SANITY CHECKS

```python
EXPECTED_COLS = [
    "Timestamp","Asset_ID","Latitude","Longitude",
    "Inventory_Level","Waiting_Time","Asset_Utilization",
    "User_Transaction_Amount","User_Purchase_Frequency",
    "Traffic_Status","Temperature","Humidity","Demand_Forecast",
    "Logistics_Delay","Logistics_Delay_Reason"
]

missing_cols = [c for c in EXPECTED_COLS if c not in df.columns]
extra_cols   = [c for c in df.columns if c not in EXPECTED_COLS]

print("Missing expected cols:", missing_cols)
print("Extra cols:", extra_cols)

# Parse datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")

# Force known categoricals
cat_cols = ["Asset_ID", "Traffic_Status", "Logistics_Delay_Reason"]
for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype("category")

# Target to int 0/1 if needed
df["Logistics_Delay"] = pd.to_numeric(df["Logistics_Delay"], errors="coerce").astype("Int64")

# Basic numeric coercion for safety
num_cols = [
    "Latitude","Longitude","Inventory_Level","Waiting_Time","Asset_Utilization",
    "User_Transaction_Amount","User_Purchase_Frequency",
    "Temperature","Humidity","Demand_Forecast"
]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Drop full duplicates (optional)
df = df.drop_duplicates().reset_index(drop=True)

print("After type fixes:", df.shape)
display(audit_df(df).head(25))
```

## 4) MISSING VALUE STRATEGY (SIMPLE + PRACTICAL)
#    - Keep it conservative for EDA
###    - For ML pipeline later, do imputation inside Pipeline

```python
# Drop rows with missing target
df = df[df["Logistics_Delay"].notna()].copy()
df["Logistics_Delay"] = df["Logistics_Delay"].astype(int)

# If lat/lon missing, you can't do geo features -> drop those rows
df = df[df["Latitude"].notna() & df["Longitude"].notna()].copy()

# For EDA: fill some key numerics with median (light touch)
for c in ["Temperature","Humidity","Inventory_Level","Waiting_Time","Demand_Forecast"]:
    if c in df.columns:
        df[c] = df[c].fillna(df[c].median())

# Categoricals: fill with "Unknown"
for c in ["Traffic_Status"]:
    if c in df.columns:
        df[c] = df[c].cat.add_categories(["Unknown"]).fillna("Unknown")

print("After missing handling:", df.shape)
display(audit_df(df).head(25))
```

# 5) LEAKAGE REMOVAL

```python
LEAKAGE_COLS = []
if "Shipment_Status" in df.columns:
    LEAKAGE_COLS.append("Shipment_Status")

if LEAKAGE_COLS:
    df = df.drop(columns=LEAKAGE_COLS)
    print("Dropped leakage columns:", LEAKAGE_COLS)
else:
    print("No leakage columns found to drop (Shipment_Status not present).")
```

## 6) FEATURE ENGINEERING (MATCHING YOUR CONTEXT)

```python
# ---- 6.1 Timestamp features
df["Hour"] = df["Timestamp"].dt.hour
df["Weekday"] = df["Timestamp"].dt.weekday  # 0=Mon ... 6=Sun
df["Is_Weekend"] = (df["Weekday"] >= 5).astype(int)

# Peak hour definition (edit to your domain)
df["Peak_Hour"] = df["Hour"].between(8, 10).astype(int) | df["Hour"].between(17, 19).astype(int)
df["Peak_Hour"] = df["Peak_Hour"].astype(int)

# ---- 6.2 Weather category derived from Temperature + Humidity (quantile-based)
temp_high = df["Temperature"].quantile(0.90)
temp_low  = df["Temperature"].quantile(0.10)
hum_high  = df["Humidity"].quantile(0.90)

def classify_weather(temp, hum):
    if (temp >= temp_high) or (hum >= hum_high):
        return "Hot"
    elif temp <= temp_low:
        return "Cold"
    else:
        return "Normal"

df["Weather"] = [classify_weather(t, h) for t, h in zip(df["Temperature"], df["Humidity"])]
df["Weather"] = df["Weather"].astype("category")

# ---- 6.3 Demand pressure = Demand_Forecast / Inventory_Level (safe divide)
df["Inventory_Level_safe"] = df["Inventory_Level"].replace(0, np.nan)
df["Demand_Pressure"] = df["Demand_Forecast"] / df["Inventory_Level_safe"]
df["Demand_Pressure"] = df["Demand_Pressure"].replace([np.inf, -np.inf], np.nan).fillna(df["Demand_Pressure"].median())

# ---- 6.4 Flags based on quantiles
wait_hi_q = df["Waiting_Time"].quantile(0.75)
df["High_Wait_Flag"] = (df["Waiting_Time"] >= wait_hi_q).astype(int)

# NOTE: Asset_Utilization is Stage-2-only for modeling,
# but for EDA you can still compute a flag.
if "Asset_Utilization" in df.columns:
    util_lo_q = df["Asset_Utilization"].quantile(0.25)
    df["Low_Util_Flag"] = (df["Asset_Utilization"] <= util_lo_q).astype(int)
else:
    df["Low_Util_Flag"] = 0

# ---- 6.5 Geo clustering via KMeans (use scaled? optional; for lat/lon usually ok)
N_CLUSTERS = 5
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE, n_init="auto")
df["Geo_Cluster"] = kmeans.fit_predict(df[["Latitude", "Longitude"]])
df["Geo_Cluster"] = df["Geo_Cluster"].astype("int64")

# ---- 6.6 Distance from central depot (Haversine)
from math import radians, cos, sin, asin, sqrt

def haversine_km(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371
    return c * r

# If you know your real depot coords, set them here:
DEPOT_LAT, DEPOT_LON = 28.61, 77.23   # <-- change to your true depot center
df["Distance_From_Center"] = df.apply(
    lambda r: haversine_km(DEPOT_LON, DEPOT_LAT, r["Longitude"], r["Latitude"]),
    axis=1
)

print("Feature engineering done.")
display(df.head())
```

## 7) QUICK TARGET OVERVIEW

```python
delay_rate = df["Logistics_Delay"].mean()
print(f"Overall delay rate: {delay_rate:.3f}")

plt.figure(figsize=(5,4))
sns.countplot(data=df, x="Logistics_Delay")
plt.title("Target Distribution: Logistics_Delay")
plt.tight_layout()
plt.show()
```

## 8) UNIVARIATE EDA (NUMERIC DISTRIBUTIONS vs TARGET)

```python
NUM_EDA = [
    "Inventory_Level","Waiting_Time","Temperature","Humidity","Demand_Forecast",
    "Demand_Pressure","Distance_From_Center"
]

for col in NUM_EDA:
    if col not in df.columns:
        continue

    plt.figure(figsize=(6,4))
    sns.kdeplot(data=df, x=col, hue="Logistics_Delay", common_norm=False)
    plt.title(f"Distribution of {col} by Delay")
    plt.tight_layout()
    plt.show()
```

## 9) CATEGORICAL EDA (DELAY RATE BY CATEGORY)

```python
def delay_rate_by(col):
    t = df.groupby(col)["Logistics_Delay"].mean().sort_values(ascending=False)
    return t

# Traffic status effect
if "Traffic_Status" in df.columns:
    print("Delay rate by Traffic_Status:")
    display(delay_rate_by("Traffic_Status"))

    plt.figure(figsize=(6,4))
    sns.barplot(x=delay_rate_by("Traffic_Status").index, y=delay_rate_by("Traffic_Status").values)
    plt.title("Delay Rate by Traffic_Status")
    plt.ylabel("Delay Rate")
    plt.tight_layout()
    plt.show()

# Weather effect
print("Delay rate by Weather:")
display(delay_rate_by("Weather"))

plt.figure(figsize=(6,4))
sns.barplot(x=delay_rate_by("Weather").index, y=delay_rate_by("Weather").values)
plt.title("Delay Rate by Weather")
plt.ylabel("Delay Rate")
plt.tight_layout()
plt.show()

# Geo cluster effect
print("Delay rate by Geo_Cluster:")
display(delay_rate_by("Geo_Cluster"))

plt.figure(figsize=(6,4))
sns.barplot(x=delay_rate_by("Geo_Cluster").index.astype(str), y=delay_rate_by("Geo_Cluster").values)
plt.title("Delay Rate by Geo_Cluster")
plt.ylabel("Delay Rate")
plt.tight_layout()
plt.show()
```

### 10) INTERACTION INSIGHTS (MATCHING YOUR FINDINGS)
##     - Traffic_Status x Cluster
##     - Weather x Cluster

```python
if "Traffic_Status" in df.columns:
    traffic_cluster = df.pivot_table(
        index="Geo_Cluster", columns="Traffic_Status", values="Logistics_Delay", aggfunc="mean"
    )
    display(traffic_cluster)

    # Ensure the data in traffic_cluster is numeric (float)
    traffic_cluster_numeric = traffic_cluster.astype(float)

    plt.figure(figsize=(10,5))
    sns.heatmap(traffic_cluster_numeric, annot=True, fmt=".2f")
    plt.title("Delay Rate: Geo_Cluster x Traffic_Status")
    plt.tight_layout()
    plt.show()

weather_cluster = df.pivot_table(
    index="Geo_Cluster", columns="Weather", values="Logistics_Delay", aggfunc="mean"
)
display(weather_cluster)

# Ensure the data in weather_cluster is numeric (float)
weather_cluster_numeric = weather_cluster.astype(float)

plt.figure(figsize=(8,5))
sns.heatmap(weather_cluster_numeric, annot=True, fmt=".2f")
plt.title("Delay Rate: Geo_Cluster x Weather")
plt.tight_layout()
plt.show()
```

## 11) CORRELATION (NUMERIC ONLY)

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# remove obvious IDs / target leakage-ish derived helper
numeric_cols = [c for c in numeric_cols if c not in ["Logistics_Delay"]]

corr = df[numeric_cols + ["Logistics_Delay"]].corr(numeric_only=True)

plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=False)
plt.title("Correlation Heatmap (Numeric)")
plt.tight_layout()
plt.show()

# Show top correlations with target
target_corr = corr["Logistics_Delay"].drop("Logistics_Delay").sort_values(key=lambda s: s.abs(), ascending=False)
print("Top correlations with Logistics_Delay (absolute):")
display(target_corr.head(15))
```
